```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\rahul\Python_libaries\Projects\Exploratory_Data_Analysis\DATA\Sample - Superstore.csv",encoding="latin-1")
print(df.head())
print(df.info())
```

```python
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

## sales and profit trend over time

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.groupby(df['Order Date'].dt.to_period("M"))[['Sales', 'Profit']].sum().plot(figsize=(12, 6), title="Monthly Sales & Profit Trend",kind='line')
plt.savefig(r'C:\Users\rahul\Python_libaries\Projects\Exploratory_Data_Analysis\IMAGES\Sales_and_Profit_Trend.png')
```

## category wise perfomance

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Category',y='Sales',estimator=sum)
plt.title('Total Sales By Category')
plt.savefig(r'C:\Users\rahul\Python_libaries\Projects\Exploratory_Data_Analysis\IMAGES\Caterory Wise Perfomance.png')
```

## Regional Performance

```python
region_profit=df.groupby('Region')['Profit'].sum().sort_values()
region_profit.plot(kind='barh',title='Profit By Regional')
plt.savefig(r'C:\Users\rahul\Python_libaries\Projects\Exploratory_Data_Analysis\IMAGES\Regional Perfomance.png')
```

## Customer Segments

```python
sns.boxenplot(data=df,x='Segment',y='Profit')
plt.title('Profit Distribution By Customer Segment')
plt.savefig(r'C:\Users\rahul\Python_libaries\Projects\Exploratory_Data_Analysis\IMAGES\Profit Distribution By Customer Segment.PNG')
```
